```python
class Patient:

  # acting as the Node class
  def __init__(self, id, name, age, bloodgroup, next, prev):
    self.id = id
    self.name = name
    self.age = age
    self.bloodgroup = bloodgroup

    self.next = next
    self.prev = prev
```
